```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import load_model
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Define your class names
class_names = ['Actinic keratosis', 'Basal cell carcinoma', 'Benign keratosis', 'melanoma']  # Replace with your actual class names

# Load your custom model
model = load_model("C:/Users/vikas/OneDrive/Desktop/Derma_GAN/CNN_with_GAN_model.h5")

# Path to your input image
image_path = "C:/Users/vikas/OneDrive/Desktop/Derma_GAN/zzzzzzzzz/Actinic keratosis/correct/ISIC_0056437.jpg"

# Load and preprocess the image
img = Image.open(image_path)
img_resized = img.resize((64, 64))  # Adjust size if your model requires different input size
img_array = image.img_to_array(img_resized)
img_array = np.expand_dims(img_array, axis=0)
img_array = img_array / 255.0  # Scale image if your model was trained on scaled images

# Make prediction
predictions = model.predict(img_array)
predicted_index = np.argmax(predictions, axis=1)[0]
predicted_class = class_names[predicted_index]
confidence = predictions[0][predicted_index]

# Display the image and prediction result
plt.imshow(img)
plt.title(f'Predicted Class: {predicted_class} ({confidence*100:.2f}%)')
plt.axis('off')
plt.show()
```
